```python
# Student Performance Predictor
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Dataset
data = {
    'study_hours': [5, 2, 3, 8, 6, 1, 4, 7, 9, 2],
    'attendance': [90, 60, 70, 95, 80, 50, 75, 85, 98, 55],
    'past_score': [80, 50, 60, 85, 78, 45, 65, 82, 90, 48],
    'result': ['Pass', 'Fail', 'Fail', 'Pass', 'Pass', 'Fail', 'Fail', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
# Encode categorical result
df['result'] = df['result'].map({'Pass': 1, 'Fail': 0})
```

```python
# Features and labels
X = df[['study_hours', 'attendance', 'past_score']]
y = df['result']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Model training
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
# Evaluation
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy*100:.2f}%")
print("Confusion Matrix:\n", cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
